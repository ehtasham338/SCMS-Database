# [dbo].[sp_Admin_GetAllServiceRequests]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Admin_GetAllServiceRequests]    Script Date: 03-Jan-2026 4:50:54 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_Admin_GetAllServiceRequests]
AS
BEGIN
    SELECT 
        sr.RequestId,
        u.FullName AS CustomerName,
        c.Model AS CarModel,
        c.NumberPlate,
        st.Name AS ServiceType,
        sr.Status,
        sr.RequestedDate,
        sr.CompletedDate,
        sa.StaffId,
        b.Amount AS BillAmount,
        b.Paid AS IsPaid,
        f.Rating,
        f.Comments
    FROM ServiceRequests sr
    INNER JOIN Users u ON sr.UserId = u.UserId
    INNER JOIN Cars c ON sr.CarId = c.CarId
    INNER JOIN ServiceTypes st ON sr.ServiceTypeId = st.ServiceTypeId
    LEFT JOIN StaffAssignments sa ON sr.RequestId = sa.RequestId
    LEFT JOIN Bills b ON sr.RequestId = b.RequestId
    LEFT JOIN Feedbacks f ON sr.RequestId = f.RequestId
    ORDER BY sr.RequestedDate DESC
END
GO
```

# [dbo].[sp_GetCompletedServices]

```sql
/****** Object:  StoredProcedure [dbo].[sp_GetCompletedServices]    Script Date: 03-Jan-2026 4:50:54 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_GetCompletedServices]
AS
BEGIN
    WITH CompletedServices AS
    (
        SELECT sr.RequestId, u.FullName AS CustomerName, sr.CarId, sr.Status, 
               b.Amount AS BillAmount, f.Rating, f.Comments AS Feedback
        FROM ServiceRequests sr
        LEFT JOIN Bills b ON sr.RequestId = b.RequestId
        LEFT JOIN Feedbacks f ON sr.RequestId = f.RequestId
        INNER JOIN Users u ON sr.UserId = u.UserId
        WHERE sr.Status = 'Completed'
    )
    SELECT *
    FROM CompletedServices;
END
GO
```

# [dbo].[sp_GetRevenueReport]

```sql
/****** Object:  StoredProcedure [dbo].[sp_GetRevenueReport]    Script Date: 03-Jan-2026 4:50:54 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_GetRevenueReport]
AS
BEGIN
    WITH Revenue AS
    (
        SELECT sr.RequestId, u.FullName, b.Amount, b.PaidDate
        FROM Bills b
        INNER JOIN ServiceRequests sr ON b.RequestId = sr.RequestId
        INNER JOIN Users u ON sr.UserId = u.UserId
        WHERE b.Paid = 1
    )
    SELECT SUM(Amount) AS TotalRevenue, COUNT(RequestId) AS TotalPaidRequests
    FROM Revenue;
END
GO
```

# [dbo].[sp_Staff_GetAssignedJobs]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Staff_GetAssignedJobs]    Script Date: 03-Jan-2026 4:50:54 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_Staff_GetAssignedJobs]
    @StaffId INT
AS
BEGIN
    SELECT 
        sr.RequestId,
        u.FullName AS CustomerName,
        c.Model AS CarModel,
        c.NumberPlate,
        st.Name AS ServiceType,
        sr.Status,
        sr.RequestedDate,
        sr.CompletedDate,
        sa.AssignedDate
    FROM StaffAssignments sa
    INNER JOIN ServiceRequests sr ON sa.RequestId = sr.RequestId
    INNER JOIN Users u ON sr.UserId = u.UserId
    INNER JOIN Cars c ON sr.CarId = c.CarId
    INNER JOIN ServiceTypes st ON sr.ServiceTypeId = st.ServiceTypeId
    WHERE sa.StaffId = @StaffId
    ORDER BY sr.RequestedDate DESC
END
GO
```

# [dbo].[sp_UpdateBillPayment]

```sql
/****** Object:  StoredProcedure [dbo].[sp_UpdateBillPayment]    Script Date: 03-Jan-2026 4:50:54 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Update bill payment
CREATE PROCEDURE [dbo].[sp_UpdateBillPayment]
    @BillId INT,
    @Paid BIT
AS
BEGIN
    UPDATE Bills
    SET Paid = @Paid,
        PaidDate = CASE WHEN @Paid = 1 THEN GETDATE() ELSE NULL END
    WHERE BillId = @BillId;
END
GO
```
